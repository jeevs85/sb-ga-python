```python

import csv
import os
import pandas as pd      
import datetime

class Budget2:
    def __init__(self, account_name):
      #intiailizes the 4 lists that are part of a transaction
      self.account_name = account_name
      self.title = []
      self.type = [] 
      self.amount_budget = []
      self.date = []  

      #headings for CSV file to be used if it's a new account
      self.headings = ["Title", "Type", "Budget", "Date"]
      #check to make sure the path doesnt exist, if new a new file is created with headings above.
      if not os.path.exists(f"./budget_{account_name}.csv"):
        with open(f"./budget_{account_name}.csv", "a+") as budget_file:
          writer = csv.DictWriter(budget_file, fieldnames=self.headings)
          writer.writeheader()
      else:
        #if account is existing values from file are read into class lists for analysis
        with open(f"./budget_{account_name}.csv", "r") as budget_file:
          
          reader = csv.DictReader(budget_file, fieldnames=self.headings)
          next(reader)
          counter = 0
          
          for row in reader:
            self.title.append(row["Title"])
            self.type.append(row["Type"])
            self.amount_budget.append(row["Budget"])
            self.date.append(row["Date"])
            counter += 1
          #run the validator to ensure all the entries in the file are valid - throw an error if not. 
          self.validator()

    
    
    def profit_calculator(self,y):
      #function calculates balance based in the account
      balance = 0
      #adds to the balance if income, takes away if expense
      if self.type[y] == "Income" or self.type[y] == "income" :
        balance += float(self.amount_budget[y])
      elif self.type[y] == "Expense" or self.type[y] == "expense":
        balance -= (self.amount_budget[y])
      return balance

    def add_entry(self):
      x= 0

      self.title.append(input("Please enter the title name"))
      self.type.append(input("Please enter the if this is an income or expense account"))
      try:
        x = float(input ("Please enter the total budget amount"))
      except ValueError:
        print("Oops!  That was no valid number...")
      self.amount_budget.append(x)
      self.date.append(input ("Please enter the date in MM-DD-YYYY"))
      self.validator()
      #potential enhancement would be to put the write to file in a seperate function and decide when best to write to file
      with open(f"./budget_{self.account_name}.csv", 'a', newline='') as file: #opening the file in write mode
       writer = csv.writer(file)
       writer.writerow([self.title[-1], self.type[-1], self.amount_budget[-1], self.date[-1]])

    def get_balance(self):
      balance = 0
      #function assumes file ha
      for y in range(len(self.amount_budget)):
        balance+=self.profit_calculator(y)
 
      print(f'Your total balance is ${balance}')   
      return balance    

    def get_file(self):
      #function displays contents of CSV using the pandas function
      data = pd.read_csv(f"./budget_{self.account_name}.csv")
      print(data)
    
    def get_profit_year (self, year) :
      #calates the profit for a given year
      balance = 0
      #splice the date to match on the year - uses the profit calculator function to calculate the balance for the year
      for y in range (len(self.date)):
        if year == self.date[y][-4:] :
          balance+=self.profit_calculator(y)

      print(f"Your profit/loss for the year {year} is {balance}")
    
    def get_profit_month(self,month, year) :
    #calcluates the profit for a given month using the month and year of the date. 
      balance = 0

      for y in range(len(self.date)):
        #splice the date to grab the year and month
        if year==self.date[y][-4:] and month==self.date[y][0:2] :
          #use the calculator function to store the balance for the month in balance variable
          balance +=self.profit_calculator(y)
     
      print(f"The profit for {month}/{year} is {balance}")
    
    def validator (self):
      #funciton validates key class variables using different methods 

      #ensure there's valid dates using the date time module
      for y in range (len(self.date)):
        try:
            datetime.datetime.strptime(self.date[y], '%m-%d-%Y')
        except ValueError:
            raise ValueError("Incorrect date format, should be MM-DD-YYYY")
      
      #ensures each transaction is either an income or expense
      for y in range (len(self.type)):
        try:
          self.type[y] != "Income"  or self.type[y] != "Expense" or self.type[y] != "income"  or self.type[y] != "expense"
        except ValueError:
          raise ValueError("Incorrect transaction type")
      for y in range (len(self.amount_budget)):
        try:
          type(self.amount_budget[y] == float)
        except ValueError: 
          raise ValueError("Incorrect Budget type")


#######################################################################################
#### Main Program
#######################################################################################
user_input = -1
#prompts user for account name
user_account = input("Please enter your account name")
#intiailizes class with the account name
acc1 = Budget2(user_account)
while user_input != "Q" or "q":
  user_input = input ("A. Add a new entry \n B. Display the total account Balance \n C. View all previous entries \n D. Profit for a year \n E Profit for a month \n Q.Exit ")
  if user_input == "A":
    acc1.add_entry()
  elif user_input =="B" :
    acc1.get_balance()
  elif user_input == "C" :
    acc1.get_file()
  elif user_input == "D" :
    acc1.get_profit_year(input("Please enter the year"))
  elif user_input =="E" :
    month = input("Please enter in the month")
    year  = input("Please enter in the year")
    acc1.get_profit_month(month, year)
  elif user_input =="Q" or user_input =="q":
    break


```
