```python
# use the playlist to write these functions 

#1. write a function that returns a list of the artists names

#2. Write a function that returns the longest rnning song

#3. WRite a function that returns the total runnting time of the playlist

playlist = [
    {
        'title': 'Bodak Yellow',
        'genre': 'Pop',
        'artiste': 'Cardi B',
        'year': '2017',
        'length': 3.25
    },
    {
        'title': 'Shake it Off',
        'genre': 'Pop',
        'artiste': 'Taylor Swift',
        'year': '2016',
        'length': 3.15
    },
    {
        'title': 'Through the Night',
        'genre': 'Pop',
        'artiste': 'Travis Scott',
        'year': '2016',
        'length': 4.50
    },
    {
        'title': 'Likes',
        'genre': 'Reggae',
        'artiste': 'Chronixx',
        'year': '2017',
        'length': 3.45
    },
    {
        'title': 'From Time',
        'genre': 'Pop',
        'artiste': 'Drake, Jhene Aiko',
        'year': '2013',
        'length': 4
    }
]
def get_LongestSong(a_dict):
    title = ''
    longest = -1
    for name in a_dict:
        if name['length'] > longest :
            longest = name['length']
            title = name['title']

    return title

def get_artist (a_dict): 
    artistname = []

    for name in a_dict:
        artistname.append(name['artiste'])
    return artistname

def get_totalRunTime (a_dict):
    total = 0

    for name in a_dict:
        total += name['length']
    return total

print(get_artist(playlist))
print(get_LongestSong(playlist) )
print(get_totalRunTime(playlist))
```
